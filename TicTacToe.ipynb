```python

import random
from IPython.display import clear_output

def full_board_check(board):
    spaces = []
    for i in range(1,10):
        spaces.append(space_check(board, i))
        
    if True in spaces:
        return False
    else:
        return True
  

def player_input():
    Marker = 'WRONG'
    while Marker != 'O' or  Marker != 'X':
        Marker = input("escoge que marca deseas O or X")
        if Marker != 'O' and  Marker != 'X':
            print('cambia tu respuesta')
        else:
            break
    return Marker


def place_marker(board, marker, position):
    board[position-1] = marker
    return 

def win_check(board,mark):
    Winner = False
    #---------------------FILAS---------------------
    if mark == mark in board[0] and mark == mark in board[1] and mark == mark in board[2]:
        Winner = True
        print(f'{mark} Ganaste en la primera fila')
    elif mark == mark in board[3] and mark == mark in board[4] and mark == mark in board[5]:
        Winner = True
        print(f'{mark} Ganaste en la segunda fila')
    elif mark == mark in board[6] and mark == mark in board[7] and  mark == mark in board[8]:
        Winner = True
        print(f'{mark} Ganaste en la tercera fila')
        #---------------------COLUMNAS---------------------    
    elif mark == mark in board[0] and mark == mark in board[3] and mark == mark in board[6]:
        Winner = True
        print(f'{mark} Ganaste en la primera Columna')
    elif mark == mark in board[1] and mark == mark in board[4] and mark == mark in board[7]:
        Winner = True
        print(f'{mark} Ganaste en la segunda Columna')
    elif mark == mark in board[2] and mark == mark in board[5] and mark == mark in board[8]:
        Winner = True
        print(f'{mark} Ganaste en la tercera Columna')
    #-----------------DIAGONALES----------------------
    elif mark == mark in board[0] and mark == mark in board[4] and mark == mark in board[8]:
        Winner = True
        print(f'{mark} Ganaste en la primera diagonal')
    elif mark == mark in board[2] and mark == mark in board[4] and mark == mark in board[6]:
        Winner = True
        print(f'{mark} Ganaste en la segunda diagonal ')
        
    else:
        Winner = False
        print('Aun no hay ganador')
    return Winner

def choose_first():
    mark = random.choice(["X","O"])
    return mark

def space_check(board, position):
    if 'X' in board[position-1] or 'O' in board[position-1]: 
        return False
    else:
        return True
def player_choice(board):
    choice = 'WRONG'
    acceptable_range = range(1,10)
    within_range = False
    while choice.isdigit() == False or within_range==False:
        choice = input(' escoge una posicion entre  (1-9)   ')
        #digit check
        if choice.isdigit() == False:
            print('no se a escogido un digito por favor vuelve a intentarlo ')
            
        if choice.isdigit() == True:
            if int(choice) in acceptable_range:
                within_range = True
            else : 
                print('lo lamento, ahora no esta en el rango deseado entre 1 y 9, vuelve a intentarlo ')
                within_range = False 
    position = int(choice)
    return position
    if space_check(board, position) == True:
        print(f"la posicion {position}  esta libre")
    else:
        print(f"la posicion {position}  no esta libre")
        
def replay():
    replay = "WRONG"
    while replay != 'yes' or  replay != 'no':
        replay = input("Quieres Volver a jugar? (yes, no)  ")
        if replay != 'yes' and  replay != 'no':
            print('cambia tu respuesta, (yes\no) ')
        else:
            if replay == 'yes':
                return True
            else:
                return False


    
```
